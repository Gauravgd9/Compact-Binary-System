```
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```

df = pd.read_csv(os.path.join('', 'data.csv'))
```

```
df.head(25)
```

```
df["q"] = df["Secondary Mass"]/df["Primary Mass"]
```

```
df
```

```
df["eta"] = df["q"]/pow((1+ df["q"]),2)
```

```
df
```

```
df["chi"] = (df["Primary Spin"] + (df["Secondary Spin"] * pow(df["q"],2)))/pow((1+df["q"]),2)
df
```

```
global Am, Bm, Cm
Am = 65.716424 
Bm = -0.96924
Cm = 5.4916553

df["mf"] = Am + (Bm * (df["eta"] ** 2)) + (Cm * df["chi"])
df
```

```
x = df["chi"]
y = df["mf"]
plt.plot(x, y) 
plt.xlabel("chi") 
plt.ylabel("mf")
plt.show()
```

```
global Aa, Ba, Ca
Aa = 3.4
Ba = 6
Ca = 9.76
df["af"] = (Aa*df["eta"]) + (Ba * df["chi"]) - (Ca * (1 - df['mf']))
df["eta_sq"] = df["eta"] ** 2
df
```

```
x = df["af"]
y = df["mf"]
plt.plot(x, y) 
plt.xlabel("af") 
plt.ylabel("mf")
plt.show()
```

```
plt.xlabel("af", c = 'blue') 
plt.ylabel("mf", c = 'blue')
plt.plot(df["Secondary Spin"]+df["Primary Mass"],df["mf"],color = 'blue')
```

```
x = df["chi"]
y = df["af"]
plt.plot(x, y) 
plt.xlabel("chi") 
plt.ylabel("af")
plt.show()
```
